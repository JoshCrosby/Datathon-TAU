## Transpose tranformation in excel

```python
TC = pd.read_excel('../data/newcomb.xlsx')
```

```python
## correlation between columns
corrcol = newcomb.corr()
corrrow = TC.corr()
```

## Build Network based on correlation

```python
corrrow
```

```python
lab = list(corrrow.columns)
linkmatrix = np.matrix(corrrow)[:][:]
linkplus = (linkmatrix>0.99985)*1.0 #-np.eye(len(lab))
linkminus = (linkmatrix<-0.9)*1.0
gp = nx.from_numpy_matrix(linkplus)
gp.name = 'Positive Correlation'
gp.to_undirected()
print nx.info(gp)
```

```python
newlab = {}
for i in range(len(lab)):
    newlab[i]=lab[i]
```

```python
ll = {}
deg =  gp.degree()
lenh = len(gp.degree())+1
while lenh>len(gp.degree()):
    lenh = len(gp.degree())
    for i in gp.degree():
        if gp.degree()[i]<=2:
            gp.remove_node(i)
            deg.pop(i)
            newlab.pop(i)
        
        
for i in gp.degree():
    if deg[i]<=1:
        gp.remove_node(i)
        deg.pop(i)
    else:
        ll[i]=lab[i]
        
print nx.info(gp)
```

# Plot Network

```python
graph_pos= nx.fruchterman_reingold_layout(gp,iterations=10)
plt.figure(1,figsize=(21,12))
nx.draw_networkx_nodes(gp, graph_pos, alpha=1, node_size =[100*v for v in deg.values()])#, node_color=g.colors)
nx.draw_networkx_edges(gp, graph_pos)
plt.title(gp.name)
nx.draw_networkx_labels(gp, graph_pos, labels=ll );

```
